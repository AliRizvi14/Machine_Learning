```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/5_one_hot_encoding/Exercise/carprices.csv')
df
```

```python
df.columns
```

```python
df_new=pd.get_dummies(df)
df_new
```

```python
x=df_new.drop('Sell Price($)',axis=1)
x
```

```python
y=df_new['Sell Price($)']
y
```

```python
plt.scatter(df['Mileage'],y)
```

```python
model=linear_model.LinearRegression()
model.fit(x,y)
```

Price of mercedez benz that is 4 yr old with mileage 45000

```python
model.predict([[45000,4,0,0,1]])
```

Price of BMW X5 that is 7 yr old with mileage 86000

```python
model.predict([[86000,7,0,1,0]])
```
